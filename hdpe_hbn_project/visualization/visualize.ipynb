```python
from ase.io import read
from pymatgen.core import Structure
from pymatviz.widgets.structure import StructureWidget

p = "/Users/sarahmatesic/Documents/GitHub/ML4MSD---Final-Project---SarahM/hdpe_hbn_project/structures/configs/repaired/config_000.xyz"

atoms = read(p)

lattice = atoms.get_cell().tolist()        # 3x3 cell
cart_coords = atoms.get_positions().tolist()  # Cartesian coordinates
species = atoms.get_chemical_symbols()

# Tell pymatgen these are cartesian coordinates
pmg_struct = Structure(lattice, species, cart_coords, coords_are_cartesian=True)

StructureWidget(pmg_struct)
```

```python
import nglview as nv
from ase.io import read

path = '/Users/sarahmatesic/Documents/GitHub/ML4MSD---Final-Project---SarahM/hdpe_hbn_project/results/config_000.traj'

# load frames (sample every 5th to keep UI responsive)
frames = read(path, index=slice(0, None, 5))
print("frames loaded:", len(frames))

# create viewer from first frame
widget = nv.show_ase(frames[0])

# attach trajectory wrapper (this is the key step)
traj = nv.ASETrajectory(frames)
widget.add_trajectory(traj)

# set representation / center camera
widget.clear_representations()
widget.add_representation('ball+stick')
widget.center()

# Now the player should exist — check and use it safely
if hasattr(widget, "player") and widget.player is not None:
    print("player found — starting playback")
    widget.player.play()
    widget.player.speed = 0.5
else:
    print("No player object found. You can still use the UI slider/play button if visible.")
    # show attributes to debug if needed
    print("widget attributes:", [a for a in dir(widget) if not a.startswith("_")])
    
# display widget (last line in the cell)
widget



```
